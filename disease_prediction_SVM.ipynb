```python
df = pd.read_csv('https://raw.githubusercontent.com/sandeeptuluri/Linear-Regression/main/dataset.csv')
df1 = pd.read_csv('https://raw.githubusercontent.com/sandeeptuluri/Linear-Regression/main/Symptom-severity.csv')
```

```python
df.isna().sum()
df.isnull().sum()

cols = df.columns
data = df[cols].values.flatten()

s = pd.Series(data)
s = s.str.strip()
s = s.values.reshape(df.shape)

df = pd.DataFrame(s, columns=df.columns)

df = df.fillna(0)
df.head()
```

```python
len(df1.Symptom)
```

```python
vals = df.values
symptoms = df1['Symptom'].unique()

for i in range(0, len(symptoms)):
  vals[vals == symptoms[i]] = df1[df1['Symptom'] == symptoms[i]]['weight'].values[0]

d = pd.DataFrame(vals, columns=cols)
df = d
df.head()
```
